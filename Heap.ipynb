```python
import math
import copy
```

```python
#SWAP Function used across this notebook.
def swap(arr_in,i,parent_index):
    tmp=arr_in[i]
    arr_in[i]=arr_in[parent_index]
    arr_in[parent_index]=tmp
```

```python
#In a heap array, there is a mathmatical relation between parent and child indices.
#Following functions return parent/child node index depending on the current index.

#index starts from 0
def parent_index(index):
    return math.floor((index-1)/2)

def left_child_index(parent_index):
    return int (  (parent_index*2) + 1  )

def right_child_index(parent_index):
    return int (  (parent_index*2) + 2  )
```

# Max Heap and Min Heap
<br/>
Heap can be a min heap or a max heap.
<br/>Heap can be in an array form or a node/tree form.
<br/> 
When in the array form, the parent/child indices are mathmatically related as given above.

<br/><br/>
Max heap - Root element is the largest. When in array form, the first element is the root element.
<br/>
Min heap - Root element is the smallest. When in array form, the first elemetn is the root element.

# Heapify
<br/>is a way of traversing the max/min heap.
<br/>
It can be upwards ( while pushing new item into heap)
<br/>
or it can be downards ( while popping the root from the heap)
<br/>
<br/>
Heapify can be done using a recursive function or a while loop.
<br/>
<br/>
Complexity of traversing once is - O (log n).
<br/>
'log n' is the height of the heap where n=number of elements in heap.

# Heapify upwards of a max-heap
<br/>ie., traverse upwards from last element in the heap to the root/first element.
<br/> It is a single traverse.

```python
#this function is used when a last element is added to an existing heap.
def recur_max_heapify_upwards(i,arr_in):
    if i == 0:
        return
    parent=parent_index(i)
    #in max heap, parent should be larger.
    #so, if parent is smaller, then swap.
    #we do not worry about the other child, as it is already an existing heap.
    if arr_in[parent] < arr_in[i]:
        swap(arr_in,i,parent)
        #now check parent with its parent.
        i=parent
        if i > 0:
            recur_max_heapify_upwards(i,arr_in)

def max_heapify_upwards_using_loop(arr_in):
    i=len(arr_in)-1

    done = False

    while done == False:
        parent=parent_index(i)
        if arr_in[parent] < arr_in[i]:
            swap(arr_in,i,parent)
            i=parent
            if i == 0:
                done=True
        else:
            done=True
```

# Heapify upwards of a min-heap
<br/>ie., traverse upwards from last node to root of a min-heap once.

```python
#First called for i=len(arr_in)-1
def recur_min_heapify_upwards(i,arr_in):
    if i == 0:
        return
    parent=parent_index(i)

    #parent value should be smaller in min heap.
    #if not, ie., if current value is smaller than parent, then swap.
    #do not worry about other child of the parent, as it is an existing heap, and the other child is greater than the parent.
    #so, if current value is smaller than parent, then current value is also smaller than other child of parent.
    if arr_in[parent] > arr_in[i]:
        swap(arr_in,i,parent)
        #now this value has gone to parent position.
        #continue the comparison/swap operation with the next parent
        #till you don't find the larger parent node or till you reach the root/head end.
        i = parent
        if i > 0 :
            recur_min_heapify_upwards(i,arr_in)

#called for i=len(arr_in)-1
def min_heapify_upwards_using_loop(i,arr_in):

    done=False
    while done == False:
        parent=parent_index(i)
        #parent value should be smaller in min heap.
        #if not, ie., if current value is smaller than parent, then swap.
        #do not worry about other child of the parent, as it is an existing heap, and the other child is greater than the parent.
        #so, if current value is smaller than parent, then current value is also smaller than other child of parent.
        if arr_in[parent] > arr_in[i]:
            swap(arr_in,i,parent)
            #now this value has gone to parent position.
            #continue the comparison/swap operation with the next parent
            #till you don't find the larger parent node or till you reach the root/head end.
            i = parent
            if i == 0:
                done=True
        else:
            done=True

```

# Heapify downwards of a max heap
<br/>ie., traverse the heap downards from the first/root element once.

```python

#First called for i=0.
#This is called when we need to remove the root node or first element of already existing heap
#and have moved the last element to first/root place and now want to move this new root node down to its right place.
def recur_max_heapify_down(i,lth,arr_in):
    left = left_child_index(i)
    right=right_child_index(i)
    largest=i

    if  left < lth and  arr_in[left] > arr_in[i] :
        largest=left
    if right < lth and arr_in[right] > arr_in[largest]:
        largest=right

    #swap the value with that of child having greater value.
    #repeat the same for the child
    if largest != i:
        swap(arr_in,i, largest)
        recur_max_heapify_down(largest,lth,arr_in)

def max_heapify_down_using_loop(arr_in):
    lth=len(arr_in)
    i = lth
    done=False
    while done == False:
        left = left_child_index(i)
        right=right_child_index(i)
        largest=i
        if  left < lth and  arr_in[left] > arr_in[i] :
            largest=left
        if right < lth and arr_in[right] > arr_in[largest]:
            largest=right

        #swap the value with that of child having greater value.
        #repeat the same for the child
        if largest != i:
            swap(i, largest, arr_in)
            i=largest
```

# Heapify downards of a min-heap
<br/>ie., traverse once downwards from head/root to last element.

```python
def recur_min_heapify_down(i,lth,arr_in):
    left = left_child_index(i)
    right=right_child_index(i)
    smallest=i

    if  left < lth and  arr_in[left] < arr_in[i] :
        smallest=left
    if right < lth and arr_in[right] < arr_in[smallest]:
        smallest=right

    #swap the value with that of child having smaller value.
    #repeat the same for the child
    if smallest != i:
        swap(arr_in,i, smallest)
        recur_min_heapify_down(smallest,lth,arr_in)


def min_heapify_down_using_loop(arr_in):
    lth=len(arr_in)
    i = lth
    done=False
    while done == False:
        left = left_child_index(i)
        right=right_child_index(i)
        smallest=i
        if  left < lth and  arr_in[left] < arr_in[i] :
            smallest=left
        if right < lth and arr_in[right] < arr_in[smallest]:
            smallest=right

        #swap the value with that of child having greater value.
        #repeat the same for the child
        if smallest != i:
            swap(i, smallest, arr_in)
            i=smallest
        else:
            done=True
        if i == (lth-1):
            done=True
```

# Push into Heap
<br/>
Alway append the element to the heap array and then traverse upwards starting from the newly added node.
<br/>
Complexity - is O(log n) for one push, where n=number_of_elements_in_heap. log n is the height of the heap.

```python
## for each entry that is being inserted - it is added at the end of the array/tree.
## And comparison/swap is made with its parent recursively till the parent is higher/equal than the inserted entry.
def push_into_max_heap(target,val):

    #insert at the end of the array, ie., at the last left node of the complete binary tree.
    target.append(val)

    start_index=len(target)-1#start from the last added node

    #now travese up from the last node and rearrage.
    recur_max_heapify_upwards(start_index,target)

    return

## And comparison/swap is made with its parent recursively till the parent is lower/equal than the inserted entry.
def push_into_min_heap(target,val):

    #insert at the end of the array, ie., at the last left node of the complete binary tree.
    target.append(val)
    start_index=len(target)-1#start from the last added node upwards.

    #now travese up from the last node and rearrage.
    recur_min_heapify_upwards(start_index,target)

    return
```

# Pop from heap
<br/>
Steps:
<br/>
1. Always the first/root node is removed.
2. Last node moves to the first position
3. Traverse the tree from the new first node downwards and rearrange the heap.


```python
## O(n log n)
# each entry is compared
#remove the head node ie., the first element of array.
# then move the last node to the first place.
# Then go down the tree to rearrange the nodes.
def pop_from_max_heap(target):
    removed_item = target[0]

    #truncate the list by removing the last empty position.
    last_item=target.pop(len(target)-1)

    if len(target) > 0:
        # move the last node value to head/first position
        target[0]=last_item

    if len(target) > 1:
        #Go down the tree to move this root node to the right place.
        recur_max_heapify_down(0,len(target),target)

    return removed_item

def pop_from_min_heap(target):
    if len(target) == 0:
        raise Exception("Nothing to pop")

    removed_item=target[0]

    #remove the last position to truncate the array by 1
    last_item=target.pop(len(target)-1)

    if len(target) > 0:

        #move the last node value to the first position.
        target[0]=last_item

    if len(target) > 1:
        #now move down the tree to push the newly added first node to the right position
        recur_min_heapify_down(0,len(target),target)

    return removed_item
```

# Build heap
Complexity is - n log n
<br/>ie., for each element's push it traverses the whole tree.



```python
def build_max_heap(arr):
    heap=[]
    for i in arr:
        push_into_max_heap(heap,i)
    return heap

def build_min_heap(arr):
    heap=[]
    for i in arr:
        push_into_min_heap(heap,i)
    return heap
```

# HEAP Sort
<br/>
Complexity - O( 2 * n log n)
<br/> - to insert n log n
<br/> - to remove n log n
<br/>
<br/>
Steps:
<br/>
1)Form a min/max heap by inserting nodes one by one
<br/>
2)Now the first element is the min/max element.
<br/>
3)Remove the first node one by one to get the ordered list.
<br/>
4)For every removal - the last node will replace the first node and 
<br/>swap occurs from top to bottom of the tree.

```python
def heap_sort_ascending(arr):

    heap=build_min_heap(arr)

    sorted=[]
    lth=len(heap)
    tmp=copy.deepcopy(heap)
    for i in range (0,lth):
        item=pop_from_min_heap(tmp)
        sorted.append(item)


    return sorted

def heap_sort_descending(arr):

    heap=build_max_heap(arr)

    sorted=[]
    lth=len(heap)
    tmp=copy.deepcopy(heap)
    for i in range (0,lth):
        item=pop_from_max_heap(tmp)
        sorted.append(item)


    return sorted
```

```python
arr=[1, 3, 5, 4, 6, 13, 10, 9, 8, 15, 17,2,7,11,12,2,5,7]
print("orig_arr={}".format(arr))

target=copy.deepcopy(arr)
target=heap_sort_ascending(target)
print("heap sorted asc = {}".format(target))

target=copy.deepcopy(arr)
target=heap_sort_descending(target)
print("heap sorted desc = {}".format(target))
```

# Build Tree/Node structure from heap array

```python
class node:
    def __init__(self, val, left=None, right=None):
        self.left=left
        self.val=val
        self.right=right
        
def build_tree_struct(heap_arr_in):
    #arr is in min/max heap format
    #create node based heap out of it
    i=1
    root_l=node(heap_arr_in[0])
    assign_node_recursive(heap_arr_in,1,root_l)
    return root_l

## O(n)
## called for every node
## recursive function
def assign_node_recursive(arr,indx,my_node):
    left_indx= (indx*2)
    if left_indx > len(arr):
        return my_node
    my_node.left=node(arr[left_indx-1])
    assign_node_recursive(arr,left_indx,my_node.left)

    right_indx= (indx*2) + 1
    if right_indx > len(arr):
        return my_node
    my_node.right=node(arr[right_indx-1])
    node.right =assign_node_recursive(arr,right_indx,my_node.right)
    
    
def height_recursive(root):
    if root:
        leftval= root.left.val if root.left else "X"
        rightval=root.right.val if root.right else 'X'
        #print("{}-{}-{}".format(leftval,root.val,rightval))

        ht_left=height_recursive(root.left)
        #print("ht_left={}".format(ht_left))
        ht_right=height_recursive(root.right)
        #print("ht_right={}".format(ht_right))

        #print ("max (ht_left,ht_right)={}".format(max(ht_left,ht_right)))
        retval= 1+max(ht_left, ht_right)
        #print ("returing retval={}".format(retval))
        #print("\n\n")
        return retval
    else:
        #print("returning 0\n\n")
        return 0

arr=[1, 3, 5, 4, 6, 13, 10]
tree_arr=copy.deepcopy(arr)
root=build_tree_struct(tree_arr)

print ( "Height of the tree by traversing = {}".format(height_recursive(root)))
print ( "Height of the tree mathmatically = {}".format(math.floor(math.log(len(arr),2)+1)))
```
